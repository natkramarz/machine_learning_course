# Lasso Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt
from sklearn import linear_model
import scipy.stats as stats
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn import model_selection
```

Rozważmy następujący zbiór punktów:

```python
true_fun = lambda X: np.cos(1.5 * np.pi * X)
n_samples=20
x = np.sort(np.random.rand(n_samples))
y = true_fun(x) + np.random.randn(n_samples) * 0.1
x=np.vstack(x)
```

# Przykład
Proszę wykonać regresję (Ridge Regression) na powyższym zbiorze danych 

```python
clf = Lasso(alpha=1.0)
clf.fit(x, y) 

x_plot = np.vstack(np.linspace(0, 1, 20))
plt.plot(x_plot, clf.predict(x_plot), color='blue',linewidth=3)
plt.plot(x, y, 'ok');
plt.show()
```

## Regresja liniowa

```python
# prepare models
models = []
predicts = []
names=[]
models.append(('LR degree 2', make_pipeline(PolynomialFeatures(2), linear_model.LinearRegression()) ))
models.append(('LR degree 20', make_pipeline(PolynomialFeatures(20), linear_model.LinearRegression()) ))

x_plot = np.vstack(np.linspace(-3, 3, 1000))
for name, model in models:
    print(name)
    model.fit(x, y)
    predicts.append(model.predict(x_plot))
    names.append(name)
    
plt.plot(x, y, 'ok');
for i in range(len(models)):
    #print(i)
    plt.plot(x_plot, predicts[i],linewidth=3,label=names[i])
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
plt.legend()    
plt.show()    
```

# Zadanie 
Proszę wykonać 
* regresję dla wielomianów o stopniu **20**
* regresję (ElasticNet Regression) dla wielomianów o stopniu **20** oraz
   * alpha = 1 
   * alpha = 10 000 
   * alpha = 0.0001

```python
models = []
models.append(('Linear Regression degree 20', make_pipeline(PolynomialFeatures(20), linear_model.LinearRegression())))
models.append(('Elastic net alpha 1', make_pipeline(PolynomialFeatures(20), ElasticNet(alpha=1))))
models.append(('Elastic net alpha 10000', make_pipeline(PolynomialFeatures(20), ElasticNet(alpha=10_000))))
models.append(('Elastic net alpha 0.0001', make_pipeline(PolynomialFeatures(20), ElasticNet(alpha=0.0001))))

names = []
predicts = []

x_plot = np.vstack(np.linspace(-3, 3, 1000))
for name, model in models:
    model.fit(x, y)
    predicts.append(model.predict(x_plot))
    names.append(name)
    
plt.plot(x, y, 'ok');
for i in range(len(models)):
    plt.plot(x_plot, predicts[i],linewidth=3,label=names[i])
    plt.xlim((0, 1))
    plt.ylim((-2, 2))
plt.legend()    
plt.show()    
```

# Zdanie 
Dobierz optymalny stopień wielomianu oraz parametr alpha.

```python
seed=123
kfold = model_selection.KFold(n_splits=5, random_state=seed, shuffle=True)

grid = model_selection.GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), ElasticNet(alpha = 0.1, random_state=seed)),
                param_grid={'polynomialfeatures__degree': [1, 2, 3, 4, 5, 20], 'elasticnet__alpha': [0.01,0.1,1,10,100,1000,10000]}, 
                cv=kfold, 
                refit=False)
grid.fit(x, y)
grid.best_params_
```

Dobierz optymalny stopień wielomianu oraz parametr <tt>alpha</tt> za pomocą metody <tt>GridSearchCV</tt> dla danych reklamowych, obejmującym sprzedaż produktów i ich budżet reklamowy w trzech różnych mediach telewizyjnych, radiu, gazetach.

```python
import pandas as pd
df_adv = pd.read_csv('Advertising.csv', index_col=0)
X = df_adv[['TV', 'radio','newspaper']]
y = df_adv['sales']
df_adv.head()
```

```python
seed=123
kfold = model_selection.KFold(n_splits=5, random_state=seed, shuffle=True)

grid = model_selection.GridSearchCV(make_pipeline(PolynomialFeatures(degree=2), ElasticNet(alpha = 0.1, random_state=seed)),
                param_grid={'polynomialfeatures__degree': [1, 2, 3, 4, 5, 20], 'elasticnet__alpha': [0.01,0.1,1,10,100,1000,10000]}, 
                cv=kfold, 
                refit=True)
grid.fit(X, y)
grid.best_params_
```
